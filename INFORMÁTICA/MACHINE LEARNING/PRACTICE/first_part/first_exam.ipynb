```python
import seaborn as sns
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import apriori, association_rules
from mlxtend.preprocessing import TransactionEncoder

music=pd.read_csv('../content/music1.csv',header=None)
music
```

Draw a histogram of the frequency of items in the dataset.

```python
music.stack().value_counts().apply(lambda x: x / music.shape[0]).plot(kind='bar')
```

Make a set of association rules for support=0.01 and confidence=0.8 add parameter num_itemsets=5 (if needed).

```python
# We must obtain the rows of the dataframe
transactions = []
for i in range(music.shape[0]):
    row = music.iloc[i].dropna().tolist()
    transactions.append(row)

# Encoder
te = TransactionEncoder()
te_ary = te.fit(transactions).transform(transactions)
df = pd.DataFrame(te_ary, columns=te.columns_)


# We must obtain the items more frequent than a threshold
frequent_itemsets = apriori(df, min_support=0.01, use_colnames=True)
frequent_itemsets['length'] = frequent_itemsets['itemsets'].apply(lambda x: len(x))


# We must obtain the rules for the frequent items
rules = association_rules(frequent_itemsets, metric='confidence', min_threshold=0.8, num_itemsets=5)
rules
```

Find the set of rules for which consequents is the set {rammstein}. Find the rule with the highest support value. Give its interpretation.

```python
rules[rules['consequents']=={'rammstein'}]
```

```python
print("By definition, we can deduce this rule is the most important of them all, or at least the one most probable to be satisfied.\n")

rules.sort_values(by='support', ascending=False).head(1)
```

Which performer/band is the most frequent in this dataset.

```python
print("The most frequent artis/band in the dataset is: ", frequent_itemsets.sort_values(by='support', ascending=False).head(1)['itemsets'])
```
