# MODULE

A Python module is a file containing Python definitions and statements. A module can define functions, classes, and variables. 

We can import the functions, and classes defined in a module to another module using the import statement in some other Python source file.

```python
import math
print(math.pi)
print(math.sqrt(85))
print(math.cos(0))
print(math.exp(1))
print(math.log(5))
```

**Import all Names**

The * symbol used with the import statement is used to import all the names from a module to a current namespace.

*from module_name import \** 

The use of * has its advantages and disadvantages. If you know exactly what you will be needing from the module, it is not recommended to use *, else do so.

```python
from math import *
# if we simply do "import math", then
# math.sqrt(16) and math.factorial() are required.
print(sqrt(16))
print(factorial(6))
```

We can rename the module while importing it using the keyword.

*import Module_name as Alias_name*

```python
import math as mt
print(mt.sqrt(16))
print(mt.factorial(6))
```

## Random Module

Python Random module generates random numbers in Python These are pseudo-random numbers means they are not truly random.

**function random()**

generates random floating numbers in the range of 0.1, and 1.0. It takes no parameters and returns values uniformly distributed between 0 and 1.

```python
import random
prob = random.random()
#on interval [0,5)
w = prob * 5
print(w)
print(int(w*10))
```

**function seed(x)**

Seed function is used to save the state of a random function, so that it can generate same random numbers on multiple executions of the code on the same machine or on different machines (for a specific seed value). The seed value is the previous value number generated by the generator. For the first time when there is no previous value, it uses current system time. 

```python
random.seed(10)
print(random.random())

print(random.choice([1, 2, 3, 4]))
print(random.randint(1, 10))
```
