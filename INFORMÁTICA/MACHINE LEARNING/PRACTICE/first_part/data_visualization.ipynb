### **Handling outliers**

Outliers are those data points that are distant from most of the similar points  in other words, we can say the outliers are entities that are different from the crowd.

Outliers cause problems when it comes to building predictive models, such as long model training times, poor accuracy, an increase in error variance, a decrease in normality, and a reduction in the power of statistical tests.

```python
import numpy as np

q1 = np.quantile(house.Price, 0.25)
q3 = np.quantile(house.Price, 0.75)
iqr = q3-q1
upper_lim = q3+(1.5*iqr)
lower_lim = q1-(1.5*iqr)
print(upper_lim)
print(lower_lim)
outliers = house[(house['Price'] > upper_lim) | (house['Price'] < lower_lim)]
outliers
```

### **Matplotlib Figure**

Matplotlib figure is a Python's object. Matplotlib figure is made up of several objects of different data types.
There are three main parts to a Matplotlib figure:

1. Figure: This is the whole region of space thats created when you create any figure. The Figure object is the overall object that contains everything else.
1. Axes: An Axes object is the object that contains the x-axis and y-axis for a 2D plot. Each Axes object corresponds to a plot or a graph. You can have more than one Axes object in a Figure.
1. Axis: An Axis object contains one of the axes, the x-axis or the y-axis for a 2D plot.

Therefore, a Matplotlib figure is a Figure object which has one or more Axes objects. Each Axes object has two or three Axis objects.

![](../images/fig3.jpg)

```python
fig = plt.figure(figsize =(7, 5), facecolor='g', edgecolor='b', linewidth=7)

# Creating new axes for the figure
"""
The add_axes method expects a list of four numbers: [left, bottom, width, height], where:

left and bottom specify the position of the bottom-left corner of the axes within the figure (as fractions of the figure width and height, 
respectively, with 0 being the far left/bottom and 1 being the far right/top).

width and height specify the dimensions of the axes as fractions of the figure size.
"""
ax = fig.add_axes([1, 1, 1, 1])

# Adding the data to be plotted
ax.scatter(house.Brick,house.Price)

# Adding title to the plot
plt.title("House prices", fontsize=25, color="yellow")

# Adding label on the y-axis
plt.ylabel('Price -Axis')

# Adding label on the x-axis
plt.xlabel('Brick -Axis')

# Setting the limit of y-axis
plt.ylim(60000, 220000)

# Adding legends
plt.legend(["Prices"])

plt.show()
```

**Swarmplot**

Draws a categorical scatterplot with non-overlapping points. A swarm plot can be drawn on its own, but it is also a good complement to a box or violin plot in cases where you want to show all observations along with some representation of the underlying distribution. Arranging the points properly requires an accurate transformation between data and point coordinates. This means that non-default axis limits must be set *before* drawing the plot.

```python
#The FacetGrid class is useful when you want to visualize the distribution of a variable or the relationship between multiple variables separately within subsets of your dataset. 
g = sns.FacetGrid(house, col="Brick")
g.map(sns.histplot, "SqFt")
```

```python
g = sns.FacetGrid(house, col="Brick",hue="Neighborhood")
g.map(sns.swarmplot, "SqFt","Price")
g.add_legend()
```

```python
titanic = sns.load_dataset("titanic")
g = sns.FacetGrid(titanic, col='pclass', hue='survived', palette="Spectral")
g = g.map(sns.swarmplot, 'sex', 'age', order=["male", "female"], hue_order=[0, 1])
g.add_legend()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
penguins = sns.load_dataset("penguins")
penguins
```

1. Check for missing data and outlier observations in the set.

```python
# Missing data

#Let's check for missing values
penguins.info()

mask = penguins.isnull()
total = mask.sum()
per = 100*mask.mean()

missing_data = pd.concat([total, per], axis=1,keys=['missing value', 'percentage'])
missing_data.sort_values(by='percentage', ascending=False, inplace=True)
missing_data
```

```python
#Let's fill the missing values
for col in penguins.columns:
    penguins[col].fillna(penguins[col].mode()[0], inplace=True)

penguins
```

```python
#Outliers

q1=np.array(0)
q3=np.array(0)
iqr=np.array(0)
upper_lim=np.array(0)
lower_lim=np.array(0)
outliers = np.array(0)
columns=['bill_length_mm', 'bill_depth_mm', 'flipper_length_mm', 'body_mass_g']
#Instead of selecting manually the numeric columns, we can use penguins.select_dtypes(include=['float','int'])

for col in columns:

    #Calculations
    q1_1d = np.quantile(penguins[col], 0.25)
    q3_1d = np.quantile(penguins[col], 0.75)
    iqr_1d = q3_1d-q1_1d
    upper_lim_1d = q3_1d+(1.5*iqr_1d)
    lower_lim_1d = q1_1d-(1.5*iqr_1d)

    #Appending to the arrays
    np.append(q1, q1_1d)
    np.append(q3, q3_1d)
    np.append(iqr, q3-q1)
    np.append(upper_lim, upper_lim_1d)
    np.append(lower_lim, lower_lim_1d)
    np.append(outliers, penguins[(penguins[col] > upper_lim_1d) | (penguins[col] < lower_lim_1d)])

outliers

# Instead of using fillna() and filling in the missing values,
# we could use np.nanquantile(penguins[column],<number>) to calculate quantiles when there are missing values
```
