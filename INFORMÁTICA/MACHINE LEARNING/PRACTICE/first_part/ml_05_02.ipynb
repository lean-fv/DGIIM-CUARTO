## Frequent itemsets

The apriori function expects data in a one-hot encoded pandas dataFrame.

```python
from mlxtend.preprocessing import TransactionEncoder

te = TransactionEncoder()
te_ary = te.fit(transactions).transform(transactions)
df = pd.DataFrame(te_ary, columns=te.columns_)
df
```

```python
from mlxtend.frequent_patterns import apriori

# min_support is the minimum support value for the itemsets returned
apriori(df, min_support=0.6)
```

```python
apriori(df, min_support=0.2)
```

```python
apriori(df, min_support=0.2, use_colnames=True)
```

```python
frequent_itemsets = apriori(df, min_support=0.2, use_colnames=True)
frequent_itemsets['length'] = frequent_itemsets['itemsets'].apply(lambda x: len(x))
frequent_itemsets
```

The advantage of working with pandas DataFrames is that we can use its convenient features to filter the results. For instance, let's assume we are only interested in itemsets of length 1 that have a support of at least 30 percent.

```python
frequent_itemsets[ (frequent_itemsets['length'] == 1) &
                   (frequent_itemsets['support'] >= 0.3) ]
```

### Frozensets

[Frozensets](https://www.w3schools.com/python/ref_func_frozenset.asp)

The frozenset() function returns an unchangeable frozenset object (which is like a set object, only unchangeable).

Note that the entries in the "itemsets" column are of type frozenset, which is built-in Python type and immutable, which makes it more efficient for certain query or comparison operations.

Since frozensets are sets, the item order does not matter.

```python
frequent_itemsets[ frequent_itemsets['itemsets'] == {'BREAD', 'MILK'} ]
```

```python
import warnings
warnings.filterwarnings('ignore')
frequent_itemsets[ frequent_itemsets['itemsets'] == frozenset(('MILK', 'BREAD')) ]
```

### Sparse Format

To save memory, you may want to represent your transaction data in the sparse format. This is especially useful if you have lots of products(items).

```python
ste_ary = te.fit(transactions).transform(transactions, sparse=True)
sparse_matrix = pd.DataFrame.sparse.from_spmatrix(ste_ary, columns=te.columns_)
sparse_matrix
```

```python
apriori(sparse_matrix, min_support=0.15, use_colnames=True, verbose=1)
```

## Association Rules from Frequent Itemsets

The generate_rules takes dataframes of frequent itemsets as produced by the apriori, fpgrowth, or fpmax functions in mlxtend.association.

The generate_rules() function allows you to

1.   specify your metric of interest
2.   the according threshold


```python
from mlxtend.frequent_patterns import association_rules

association_rules(frequent_itemsets, metric="confidence", min_threshold=0.4, num_itemsets=len(frequent_itemsets))
```

For the rule $antecedents \Rightarrow consequents$
$$
\left(A\Rightarrow C\right)
$$
we have



*  the support is used to measure the abundance or frequency (often interpreted as significance or importance) of an itemset in a database.
$$
range:[0,1]
$$
*  the confidence of a rule A->C is the probability of seeing the consequent in a transaction given that it also contains the antecedent.
$$
range:[0,1]
$$
*  the lift metric is commonly used to measure how much more often the antecedent and consequent of a rule A->C occur together than we would expect if they were statistically independent. If A and C are independent, the Lift score will be exactly 1.
$$
range:[0,\infty]
$$
* the leverage computes the difference between the observed frequency of A and C appearing together and the frequency that would be expected if A and C were independent. A leverage value of 0 indicates independence.
$$
range:[-1,1]
$$
*  A high conviction value means that the consequent is highly depending on the antecedent. If items are independent, the conviction is 1.
$$
range:[0,\infty]
$$
*  the zhangs_metric measures both association and dissociation. Value ranges between -1 and 1. A positive value (>0) indicates Association and negative value indicated dissociation.
$$
range:[-1,1]
$$


```python
rules=association_rules(frequent_itemsets, metric="confidence", min_threshold=0.14, num_itemsets=len(frequent_itemsets))
rules = rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']]
rules
```

```python
rules[ (rules['confidence'] > 0.35) &
       (rules['lift'] > 1.2) ]
```

```python
rules[rules['antecedents'] == {'MILK'}]
```

```python
rules[rules['antecedents'] == {'BREAD'}]
```

Rules removal

```python
rules2=association_rules(frequent_itemsets, metric="confidence", min_threshold=0.04, num_itemsets=len(frequent_itemsets))
rules2
```

```python
antecedent_sele = rules2['antecedents'] == frozenset({'BREAD'})
consequent_sele = rules2['consequents'] == frozenset({'MILK'})
final_sele = (antecedent_sele & consequent_sele)

final_sele
```

```python
#Provides the rows that don't satisfy the condition
rules2.loc[ ~final_sele ]
```

```python
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import apriori, association_rules
from mlxtend.preprocessing import TransactionEncoder

groceries=pd.read_csv("../content/groceries - groceries.csv")
groceries=groceries[groceries.columns[1:]]
groceries
```

From the datasets in the Data folder, choose one dataset and find association rules in it (minimum 30 rules). Give an interpretation of the rule with the highest support value.

```python
# We calculate the support of each item
support=groceries.stack().value_counts().apply(lambda x: x / groceries.shape[0])

transactions = []
for i in range(groceries.shape[0]):
    #iloc transposes the rows, so dropna() removes the colums
    row = groceries.iloc[i].dropna().tolist()
    transactions.append(row)

te = TransactionEncoder()
te_ary = te.fit(transactions).transform(transactions)
df = pd.DataFrame(te_ary, columns=te.columns_)

frequent_itemsets = apriori(df, min_support=0.01, use_colnames=True)
frequent_itemsets['length'] = frequent_itemsets['itemsets'].apply(lambda x: len(x))

rules = association_rules(frequent_itemsets, metric='support', min_threshold=0.01, num_itemsets=len(frequent_itemsets))
a=rules.sort_values(by='support', ascending=False).head(1)
a
```
